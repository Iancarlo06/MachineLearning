```python
from pandas import read_csv,to_datetime,DataFrame


def get_inputs(data: DataFrame,
               window: int) -> tuple:
    x = list()
    y = list()
    size = len(data)
    for i in range(size-window-1):
        j = i+window
        xi = data.iloc[i:j]
        xi = xi.to_numpy()
        yi = data.iloc[j:j+5]
        yi = yi.to_numpy()
        yi = yi.flatten()
        x.append(xi)
        y.append(yi)
    return x, y




data = read_csv(
    "/content/drive/MyDrive/4.5_month.csv"
)
data["time"] = to_datetime(
    data["time"]
)
data["updated"] = to_datetime(
    data["updated"]
)
data["seg"] = data["updated"]-data["time"]
data["seg"] = data["seg"].apply(
    lambda time:
    time.seconds
)

data["longitude"] = data["longitude"]/180
data["latitude"] = data["latitude"]/180
data["mag"] = data["mag"]/10
data["seg"] = data["seg"]/100000
data = data[[
    "latitude",
    "longitude",
    "depth",
    "mag",
    # "seg"
]]
x, y = get_inputs(
    data,
    3
)
```
