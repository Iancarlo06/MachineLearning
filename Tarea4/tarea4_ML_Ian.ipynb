```python
import pandas as pd
import os
import shutil
```

```python
dir = 'sample_labels.csv'
data = pd.read_csv(dir)
```

```python
def indicate_folder(data): 
    size = len(data)
    train_index = int(0.6*size)
    val_index = int(0.2*size)
    val_index += train_index 
    folders = list()
    for index in range(size): 
        label = "Test"
        if index < train_index:
            label = 'Train'
        elif index < val_index:
            label = 'Validation'
        folders.append(label)
    data["Folder"] = folders
    return data
```

```python
data['Label'] = data['Finding Labels'].apply(
    lambda value:
    1 if 'Pneumonia' in value else 0
)
columns = data.columns
columns = [
    columns[0], 
    columns[-1]
]
data = data[columns]
```

```python
zeros = pd.DataFrame(
    data[data['Label'] == 0]
)
ones = pd.DataFrame(
    data[data['Label'] == 1]
)
zeros = indicate_folder(zeros)
ones = indicate_folder(ones)
data = pd.concat([zeros,ones])
data
```

```python
for index in data.index:
    filename, label, folder = list(data.loc[index])
    if filename in os.listdir(file_ori):
        file_path_old = ''.join([file_ori, filename])
        path = ''.join([cat_path_ori, folder])
        if label: 
            path = ''.join([path, '\\Pneumonia'])
        else: 
            path = ''.join([path, '\\No_Pneumonia'])
        file_path_new = ''.join([path, '\\', filename])
        shutil.move(file_path_old, file_path_new)
```
